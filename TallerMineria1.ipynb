#Separadores

```python
#separar Fecha emision licencia
df2["Fecha emision licencia"] = df2.apply(
    lambda x:
          str(x["Fecha emision licencia"]).split(sep = 't'),
          axis = 1
)
df2["Año emision licencia"] = df2.apply(
    lambda x:
        str(x["Fecha emision licencia"])[2:6],
        axis = 1
)
df2["Año emision licencia"] = df2.apply(
        lambda x:
            x["Año emision licencia"] if ((str(x["Año emision licencia"][0])).isnumeric() & (str(x["Año emision licencia"][1])).isnumeric()) else "Quitar"
            , axis = 1
    )


df2["Trimestre emision licencia"] = df2.apply(
    lambda x:
        str(x["Fecha emision licencia"])[-3:-2],
        axis = 1
)

df2 = df2.drop(columns = "Fecha emision licencia")
```

```python
#separar Fecha inicio reposo
df2["Fecha inicio reposo"] = df2.apply(
    lambda x:
          str(x["Fecha inicio reposo"]).split(sep = 't'),
          axis = 1
)
df2["Año emision reposo"] = df2.apply(
    lambda x:
        str(x["Fecha inicio reposo"])[2:6],
        axis = 1
)
df2["Año emision reposo"] = df2.apply(
        lambda x:
            x["Año emision reposo"] if ((str(x["Año emision reposo"][0])).isnumeric() & (str(x["Año emision reposo"][1])).isnumeric()) else "Quitar"
            , axis = 1
    )


df2["Trimestre emision reposo"] = df2.apply(
    lambda x:
        str(x["Fecha inicio reposo"])[-3:-2],
        axis = 1
)

df2 = df2.drop(columns = "Fecha inicio reposo")
```

```python
#separar Fecha recepcion isapre
df2["Fecha recepcion isapre"] = df2.apply(
    lambda x:
          str(x["Fecha recepcion isapre"]).split(sep = 't'),
          axis = 1
)
df2["Año recepcion isapre"] = df2.apply(
    lambda x:
        str(x["Fecha recepcion isapre"])[2:6],
        axis = 1
)
df2["Año recepcion isapre"] = df2.apply(
        lambda x:
            x["Año recepcion isapre"] if ((str(x["Año recepcion isapre"][0])).isnumeric() & (str(x["Año recepcion isapre"][1])).isnumeric()) else "Quitar"
            , axis = 1
    )


df2["Trimestre recepcion isapre"] = df2.apply(
    lambda x:
        str(x["Fecha recepcion isapre"])[-3:-2],
        axis = 1
)

df2 = df2.drop(columns = "Fecha recepcion isapre")
```

```python
#separar Fecha recepcion empleador
df2["Fecha recepcion empleador"] = df2.apply(
    lambda x:
          str(x["Fecha recepcion empleador"]).split(sep = 't'),
          axis = 1
)
df2["Año recepcion empleador"] = df2.apply(
    lambda x:
        str(x["Fecha recepcion empleador"])[2:6],
        axis = 1
)
df2["Año recepcion empleador"] = df2.apply(
        lambda x:
            x["Año recepcion empleador"] if ((str(x["Año recepcion empleador"][0])).isnumeric() & (str(x["Año recepcion empleador"][1])).isnumeric()) else "Quitar"
            , axis = 1
    )


df2["Trimestre recepcion empleador"] = df2.apply(
    lambda x:
        str(x["Fecha recepcion empleador"])[-3:-2],
        axis = 1
)

df2 = df2.drop(columns = "Fecha recepcion empleador")
```

```python
#Nuevo DataFrame
df2.info()
```

#Convertir object a int

```python
#Eliminacion de filas con datos erroneos
df2.drop(df2.index[df2['Año recepcion empleador'] == 3000], inplace = True)
df2.drop(df2.index[df2['Año recepcion empleador'] == 1800], inplace = True)
df2.drop(df2.index[df2['Año recepcion empleador'] == 2023], inplace = True)
```

```python
#Codigo para convertir los object a int
df2['Año emision licencia'] = df2['Año emision licencia'].astype(str).astype(int)
df2['Año emision reposo'] = df2['Año emision reposo'].astype(str).astype(int)
df2['Trimestre emision licencia'] = df2['Trimestre emision licencia'].astype(str).astype(int)
df2['Trimestre emision reposo'] = df2['Trimestre emision reposo'].astype(str).astype(int)
df2['Año recepcion isapre'] = df2['Año recepcion isapre'].astype(str).astype(int)
df2['Trimestre recepcion isapre'] = df2['Trimestre recepcion isapre'].astype(str).astype(int)
df2['Año recepcion empleador'] = df2['Año recepcion empleador'].astype(str).astype(int)
df2['Trimestre recepcion empleador'] = df2['Trimestre recepcion empleador'].astype(str).astype(int)
```
